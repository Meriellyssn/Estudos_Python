```python
#Criando um arquivo CSV

import csv
dados = [
    ['nome', 'idade', 'cidade'],
    ['Ana', 25, 'São Paulo'],
    ['Carlos', 30, 'Rio de Janeiro'],
    ['Maria', 28, 'Belo Horizonte']
]

with open('pessoas.csv', 'w', newline='') as arquivo:
    writer = csv.writer(arquivo)
    writer.writerows(dados)
```

API (Application Programming Interface) <br>

GET -> Solicita e recupera dados de um recurso específico. Não modifica o estado do servidor.<br>
POST -> Envia novas informações ao servidor para criar um recurso ou processar dados.

Integração com Banco de Dados

```python
#Criando ou conectando ao Banco de Dados

import sqlite3

conexao = sqlite3.connect('dados.db')
cursor = conexao.cursor()
cursor.execute("""
    CREATE TABLE usuario (
        nome TEXT,
        idade INTEGER
)
""")

conexao.commit()
```

```python
#Inserindo os Dados

import sqlite3
conexao = sqlite3.connect('dados.db')
cursor = conexao.cursor()
cursor.execute(
    'INSERT INTO usuario VALUES (?, ?)',
    ('Ana', 25)
)
conexao.commit()
```

```python
#Lendo os Dados
conexao = sqlite3.connect('dados.db')
cursor = conexao.cursor()
cursor.execute('SELECT * FROM usuario')
dados = cursor.fetchall()
print(dados)
```
