```python
# Listagem 5.5 – Imprimindo de 1 a 3 com while
from time import sleep
x = 10
while x >= 1:
    print(x)
    sleep(1)
    x -= 1
print(x)
print('Fogo!!!!')
```

```python
#Listagem 5.6 – Impressão de 1 até um número digitado pelo usuário
fim = int(input('Digite o último número a imprimir: '))
x = 1
while x <= fim:
    print(x)
    x += 1
```
