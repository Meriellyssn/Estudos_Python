```python
valor_pizza = 45
quant = 3
entrega = 7
total = valor_pizza * quant + entrega
pessoas = 3
dividir_conta = total / pessoas
print(f'O total do pedido é de R${total:.2f} reais.')
print(f'O valor de R${total:.2f}, dividido para {pessoas} pessoas,'
      f'cada uma pagará o valor de R${dividir_conta:.2f} reais.')
```

```python
valor_pizza = 45
quant = 5
entrega = 7
distancia = 8
total = valor_pizza * quant + entrega
frete_gratis = total >= 200 or distancia <= 3
print(f'O total do pedido é de R${total:.2f} reais.')
print(f'Cliente tem direito a frete grátis? {frete_gratis}')
```
