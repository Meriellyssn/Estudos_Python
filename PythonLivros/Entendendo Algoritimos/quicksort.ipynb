Código Comentado

```python
def quicksort(array):
    # CASO BASE: Se a lista estiver vazia ou tiver apenas 1 elemento,
    # ela já está ordenada por natureza. Então, basta retorná-la.
    if len(array) < 2:
        return array
    else:
        # CASO RECURSIVO: Escolhe o primeiro elemento da lista como o "Pivot" (pivô)
        pivo = array[0]

        # Cria uma nova lista com todos os elementos menores ou iguais ao pivô.
        # Ele pula o pivô olhando a partir do índice 1 (array[1:])
        menores = [i for i in array[1:] if i <= pivo]

        # Cria uma nova lista com todos os elementos maiores que o pivô.
        maiores = [i for i in array[1:] if i > pivo]

        # A MAGIA DA RECURSIVIDADE:
        # Ele chama a função 'quicksort' novamente para ordenar a lista dos menores,
        # faz o mesmo para a lista dos maiores, e junta tudo com o pivô no meio.
        return quicksort(menores) + [pivo] + quicksort(maiores)

# Executa o algoritmo passando uma lista desordenada e imprime o resultado na tela
print(quicksort([10, 5, 2, 3]))
```

Exercício 4.2: Escrever uma função recursiva que conte o número de itens em uma lista
A lógica para contar elementos é muito parecida com a de somá-los.
Caso-base: Uma lista vazia [] possui $0$ elementos.
Caso recursivo: O tamanho da lista é $1$ (representando o elemento atual) mais a contagem do restante da lista.

```python
def contar_elementos(lista):
    # Caso-base: se a lista está vazia, o total é 0
    if not lista:
        return 0

    # Caso recursivo: 1 + quantidade de elementos do resto da lista
    return 1 + contar_elementos(lista[1:])

# Como testar:
nomes = ['Alice', 'Bob', 'Charlie', 'David']
print(f'A lista possue {contar_elementos(nomes)} nomes.')
```

Exercício 4.3: Encontrar o valor mais alto em uma lista de forma recursiva
Para encontrar o valor máximo usando recursão, precisamos comparar elementos passo a passo.
Caso-base: * Se a lista tiver apenas um elemento, ele é obrigatoriamente o maior valor da lista.
Se a lista estiver vazia, não há valor máximo (retornamos None).
Caso recursivo: Comparamos o primeiro elemento da lista com o "maior elemento do resto da lista". Quem for maior, ganha!

```python
def maximo_recursivo(lista):
    #Tratamento de erro para lista vazia
    if len(lista) == 0:
        return None
    #Caso-base: lista com apenas 1 elemento
    if len(lista) == 1:
        return lista[0]
    #Caso recursivo: encontra o maior valor do restante da lista
    sub_maximo = maximo_recursivo(lista[1:])

    #Compara o primeiro elemento com o maior valor encontrado no resto
    if lista[0] > sub_maximo:
        return lista[0]
    else:
        return sub_maximo

#Como testa:
valores = [5, 20, 15, 30, 10]
print(f'O maior valor da lista é: {maximo_recursivo(valores)}')
```

4.5 Imprimir o valor de cada elemento em um array

Aqui, o algoritmo precisa passar por cada um dos elementos do array, um por um, e realizar a operação de impressão (print).

```python
def imprimir_elementos(arr):
    for elemento in arr:
        print(elemento)
```

Análise de Tempo:

Se o array tiver $10$ elementos, o programa fará $10$ impressões. Se tiver $1.000$ elementos, fará $1.000$ impressões. O tempo de execução cresce de forma perfeitamente direta e proporcional ao número de elementos.

Tempo em Notação Big O: $O(n)$

Classificação: Tempo Linear.

4.6 Duplicar o valor de cada elemento em um array

Nesta operação, nós também percorremos o array inteiro do início ao fim. Para cada elemento, fazemos uma multiplicação básica (que leva um tempo constante, muito rápido) e salvamos o novo valor.

```python
def duplicar_elementos(arr):
    for i in range(len(arr)):
        arr[i] *= 2
```

Análise de Tempo:

Assim como no exercício anterior, se o array crescer, o número de operações de duplicação cresce exatamente na mesma proporção. Mesmo que a operação de multiplicar por $2$ seja ligeiramente diferente de imprimir, a taxa de crescimento em relação a $n$ continua sendo a mesma.

Tempo em Notação Big O: $O(n)$

Classificação: Tempo Linear.

4.8 Criar uma tabela de multiplicação com todos os elementos do array

Este exercício é o mais interessante. Se o array for [2, 3, 7, 8, 10] (tamanho $n = 5$):

Primeiro, multiplicamos todos os $5$ elementos por $2$ ($5$ operações).

Depois, multiplicamos todos os $5$ elementos por $3$ ($5$ operações).

Depois, por $7$ ($5$ operações), e assim por diante.

No total, realizamos $5$ rodadas de $5$ multiplicações, resultando em $25$ operações no total ($5 \times 5$).

```python
def tabela_multiplicacao(arr):
    for x in arr: # Roda n vezes
        for y in arr: # Roda n vezes para cada ciclo do loop externo
            print(x * y) # Executado n * n vezes
```

Análise de Tempo:

Como temos um loop dentro de outro loop (loops aninhados) percorrendo o mesmo array de tamanho $n$, o número total de operações será sempre $n \times n$.

Se $n = 10$, faremos $100$ operações ($10^2$).

Se $n = 100$, faremos $10.000$ operações ($100^2$).

O tempo de execução cresce de forma quadrática.

Tempo em Notação Big O: $O(n^2)$

Classificação: Tempo Quadrático.
