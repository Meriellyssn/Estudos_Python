```python
votaram = {}
def verificar_eleitor(nome):
    if votaram.get(nome):
        print('Já Votou!')
    else:
        votaram[nome] = True
        print('Deixar votar!')

verificar_eleitor('tom')
verificar_eleitor('mike')
verificar_eleitor('tom')
```
