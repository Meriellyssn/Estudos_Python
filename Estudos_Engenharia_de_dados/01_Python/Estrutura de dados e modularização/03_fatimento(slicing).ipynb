A Regra de Ouro do Fatiamento<br>
Para fatiar uma lista, usamos os colchetes [ ] (como fazíamos para acessar um item só), mas adicionamos dois pontos : para definir um intervalo. A fórmula completa tem três partes:

```python
# minha_lista[inicio : fim : passo]
```

inicio: O índice de onde a fatia começa (incluído).<br>

fim: O índice onde a fatia termina (excluído - o Python pega até um item antes desse número).<br>

passo: De quantos em quantos itens ele vai pular (opcional, o padrão é 1).

```python
# Índices:        0    1    2    3    4    5
vendas_meses = [150, 200, 180, 220, 250, 300]
```
