1. Métodos para Adicionar Dados<br>
.append(item): Este é provavelmente o método que você mais vai usar. Ele adiciona um único item ao final da lista.

```python
tabela_banco = ['cliente', 'vendas']
tabela_banco.append('produtos')
print(tabela_banco)
```

.extend(outra_lista): Diferente do append (que adiciona um item só), o extend pega todos os itens de uma outra lista e junta no final da sua lista atual. É perfeito para juntar lotes de dados.

```python
lote_01 = [101, 102]
lote_02 = [103, 104]
lote_01.extend(lote_02)
print(lote_01)
```

.insert(indice, item): Se você não quiser colocar o dado no final, mas sim em uma "gaveta" específica, use o insert. Você diz a posição (índice) e o que quer colocar lá.

```python
# Colocando "falha_rede" na posição 0 (primeiro lugar)
erros = ['timeout', 'perda_pacote']
erros.insert(0, 'falha_rede')
print(erros)
```

2. Métodos para Remover Dados<br>
.remove(valor): Remove a primeira ocorrência de um valor específico que você disser. Se o valor não existir, o Python dará um erro.

```python
status = ['ok', 'erro', 'ok', 'pendente']
status.remove('ok')
print(status)
```

.pop(indice): Este método tira um item de uma posição específica e o "entrega" para você. Se você usar o .pop() sem colocar nenhum número dentro dos parênteses, ele removerá e entregará o último item da lista.

```python
fila_processamento = ['job_A', 'job_B', 'job_C']
job_atual = fila_processamento.pop(0)
print(f'Processando agora: {job_atual}')
print(f'Fila restante: {fila_processamento}')
```

3. Métodos para Organizar e Consultar<br>
.sort(): Organiza a lista. Se for de números, coloca em ordem crescente. Se for de textos, coloca em ordem alfabética.

```python
tempo_ms = [45, 12, 89, 2]
tempo_ms.sort()
print(tempo_ms)
```

Dica Bônus - Função len(lista): Isso não é um método (repare que não usamos o ponto .), mas uma função nativa do Python extremamente útil. Ela retorna o tamanho da lista, ou seja, quantos itens existem lá dentro.

```python
tabelas_banco = [['maria'],['joão', 25], ['ana']]

total_tabelas = len(tabelas_banco)
print(f"Temos {total_tabelas} tabelas no banco de dados.")
print(tabelas_banco)
```
