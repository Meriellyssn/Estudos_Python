```python
# Criamos um array fixo com 10 posições vazias
tamanho_tabela = 10
tabela_memoria = [None] * tamanho_tabela

def inserir_na_hash(chave, valor):
    # O hash() converte o texto em um número inteiro gigante.
    # O operador % (resto da divisão) garante que o índice caiba no nosso array (0 a 9).
    indice = hash(chave) % tamanho_tabela

    # Armazena o valor direto na posição calculada
    tabela_memoria[indice] = valor
    print(f"Chave '{chave}' gerou o índice {indice} na memória.")

# Testando a nossa função hash manual
inserir_na_hash("Maçã", 4.50)
inserir_na_hash("Banana", 3.20)
```
